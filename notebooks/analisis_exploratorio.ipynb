```python
##FUNCION PARA NORMALIZAR HOJAS
def estandarizar(df):

    # normalizar columnas
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(' ', '_')
        .str.replace('á','a')
        .str.replace('é','e')
        .str.replace('í','i')
        .str.replace('ó','o')
        .str.replace('ú','u')
    )

    # unificar títulos
    if 'titulo_documento' in df.columns:
        df = df.rename(columns={'titulo_documento': 'titulo'})

    if 'titulo_de_documento' in df.columns:
        df = df.rename(columns={'titulo_de_documento': 'titulo'})

    # unificar if_5y
    if 'if5_y' in df.columns:
        df = df.rename(columns={'if5_y': 'if_5y'})

    if '5yif' in df.columns:
        df = df.rename(columns={'5yif': 'if_5y'})

    # unificar autor bool
    if 'primer_autor_inta_(si/no)' in df.columns:
        df = df.rename(columns={'primer_autor_inta_(si/no)': 'primer_autor_inta'})

    if 'primer_autor_inta_(si,_no)' in df.columns:
        df = df.rename(columns={'primer_autor_inta_(si,_no)': 'primer_autor_inta'})

    return df
```

## Post-limpieza estructural para análisis SQL y Power BI

Durante la revisión posterior a la carga en SQLite se identificaron inconsistencias estructurales y de normalización que podían afectar las consultas analíticas y las visualizaciones.

Las principales correcciones realizadas en esta etapa son:

- eliminación de columnas auxiliares generadas por inconsistencias históricas en los nombres de columnas
- consolidación de variantes como `if2022` e `if2023` dentro de la columna analítica `if`
- normalización de valores categóricos en `primer_autor_inta`
- reemplazo de valores numéricos no válidos en campos textuales por valores nulos
- preparación de una tabla final más consistente para SQL y Power BI

Estas correcciones permiten mantener una capa de datos limpia antes de construir métricas, consultas y dashboards.

```python
import numpy as np

# -----------------------------
# 1. Consolidar columnas IF mal nombradas
# -----------------------------
columnas_if_aux = [col for col in df_total.columns if col.lower() in ["if2022", "if2023"]]

if "if" not in df_total.columns:
    df_total["if"] = np.nan

for col in columnas_if_aux:
    df_total["if"] = df_total["if"].fillna(df_total[col])

# Eliminar columnas IF auxiliares
df_total = df_total.drop(columns=columnas_if_aux, errors="ignore")


# -----------------------------
# 2. Eliminar columna duplicada de primer autor
# -----------------------------
columnas_a_eliminar = [
    "primer_autor_inta(si,_no)",
    "primer_autor_inta (si,_no)",
    "primer_autor_inta(si, no)",
    "primer_autor_inta (si, no)"
]

df_total = df_total.drop(columns=columnas_a_eliminar, errors="ignore")


# -----------------------------
# 3. Normalizar primer_autor_inta
# -----------------------------
if "primer_autor_inta" in df_total.columns:
    df_total["primer_autor_inta"] = (
        df_total["primer_autor_inta"]
        .astype("string")
        .str.strip()
        .str.lower()
        .str.replace("í", "i", regex=False)
    )

    df_total["primer_autor_inta"] = df_total["primer_autor_inta"].replace({
        "si": "si",
        "sí": "si",
        "s": "si",
        "yes": "si",
        "no": "no",
        "n": "no"
    })

    df_total["primer_autor_inta"] = df_total["primer_autor_inta"].replace({
        "nan": pd.NA,
        "none": pd.NA,
        "": pd.NA,
        "0": pd.NA
    })


# -----------------------------
# 4. Reemplazar ceros inválidos en columnas textuales
# -----------------------------
columnas_texto_sensibles = [
    "academicos_inta_correspondientes",
    "coautores_inta",
    "autores",
    "titulo",
    "title"
]

for col in columnas_texto_sensibles:
    if col in df_total.columns:
        df_total[col] = df_total[col].replace({
            0: pd.NA,
            "0": pd.NA,
            "nan": pd.NA,
            "None": pd.NA,
            "": pd.NA
        })


# -----------------------------
# 5. Auditoría rápida
# -----------------------------
print("Columnas IF auxiliares eliminadas:", columnas_if_aux)
print("Columnas actuales:", df_total.shape)
print("\nValores primer_autor_inta:")
if "primer_autor_inta" in df_total.columns:
    print(df_total["primer_autor_inta"].value_counts(dropna=False))
```

## Ajustes finales de consistencia estructural

Previo a la exportación definitiva del dataset, se realiza una última auditoría orientada a detectar inconsistencias remanentes observadas durante la carga en entorno SQL.

En esta etapa se corrigen:

- columnas auxiliares o duplicadas no relevantes para análisis
- variables mal consolidadas entre años
- normalización de campos categóricos
- reemplazo de valores numéricos inválidos en columnas textuales

Con esto se asegura una tabla final estable para explotación analítica en SQLite y Power BI.

```python
import pandas as pd
import numpy as np
import re
import unicodedata

# =============================
# POST-LIMPIEZA FINAL V2
# =============================

def normalizar_nombre_columna(col):
    col = str(col).strip().lower()
    col = unicodedata.normalize("NFKD", col).encode("ascii", "ignore").decode("utf-8")
    col = re.sub(r"\s+", "_", col)
    col = re.sub(r"[^a-z0-9_]", "_", col)
    col = re.sub(r"_+", "_", col)
    return col.strip("_")


# -----------------------------
# 1. Consolidar columnas IF mal nombradas hacia "if"
# -----------------------------
cols_if_aux = [
    col for col in df_total.columns
    if re.fullmatch(r"if_?\d{4}", normalizar_nombre_columna(col))
]

if "if" not in df_total.columns:
    df_total["if"] = pd.NA

for col in cols_if_aux:
    df_total["if"] = df_total["if"].fillna(df_total[col])

df_total = df_total.drop(columns=cols_if_aux, errors="ignore")


# -----------------------------
# 2. Eliminar columnas basura / duplicadas
# -----------------------------
cols_a_eliminar = []

for col in df_total.columns:
    col_norm = normalizar_nombre_columna(col)

    # Eliminar revista
    if col_norm == "revista":
        cols_a_eliminar.append(col)

    # Eliminar columnas tipo primer_autor_inta_(si,_no)
    if "primer_autor_inta" in col_norm and "si" in col_norm and "no" in col_norm:
        cols_a_eliminar.append(col)

df_total = df_total.drop(columns=list(set(cols_a_eliminar)), errors="ignore")


# -----------------------------
# 3. Normalizar primer_autor_inta
# -----------------------------
if "primer_autor_inta" in df_total.columns:
    df_total["primer_autor_inta"] = (
        df_total["primer_autor_inta"]
        .astype("string")
        .str.strip()
        .str.lower()
        .str.replace("í", "i", regex=False)
    )

    df_total["primer_autor_inta"] = df_total["primer_autor_inta"].replace({
        "sí": "si",
        "si": "si",
        "s": "si",
        "yes": "si",
        "y": "si",
        "no": "no",
        "n": "no",
        "0": pd.NA,
        "0.0": pd.NA,
        "nan": pd.NA,
        "none": pd.NA,
        "": pd.NA
    })


# -----------------------------
# 4. Reemplazar 0 inválidos en columnas textuales
# -----------------------------
columnas_texto_a_limpiar = [
    "academicos_inta_correspondiente",
    "academicos_inta_correspondientes",
    "coautor_inta",
    "coautores_inta",
    "autor_inta",
    "autores_inta"
]

for col in df_total.columns:
    col_norm = normalizar_nombre_columna(col)

    if col_norm in columnas_texto_a_limpiar:
        df_total[col] = df_total[col].replace({
            0: pd.NA,
            0.0: pd.NA,
            "0": pd.NA,
            "0.0": pd.NA,
            "nan": pd.NA,
            "none": pd.NA,
            "None": pd.NA,
            "": pd.NA
        })


# -----------------------------
# 5. Auditoría final
# -----------------------------
print("Columnas IF eliminadas:")
print(cols_if_aux)

print("\nColumnas eliminadas:")
print(list(set(cols_a_eliminar)))

print("\nShape final:")
print(df_total.shape)

print("\nColumnas sospechosas restantes:")
for col in df_total.columns:
    col_norm = normalizar_nombre_columna(col)
    if "if" in col_norm or "primer_autor_inta" in col_norm or "revista" in col_norm:
        print("-", col)

if "primer_autor_inta" in df_total.columns:
    print("\nValores primer_autor_inta:")
    print(df_total["primer_autor_inta"].value_counts(dropna=False))
```

Tras una nueva validación del dataset en entorno SQLite se detectaron dos inconsistencias remanentes que podían seguir afectando la explotación analítica:

- permanencia de una columna duplicada sin contenido asociada a `primer_autor_inta`
- persistencia de valores `0` en campos textuales vinculados a académico INTA correspondiente

Debido a que ambas situaciones generan ruido estructural y categorías no válidas dentro de consultas SQL y futuras visualizaciones, se realiza una corrección complementaria orientada a:

- depurar columnas redundantes no analíticas
- asegurar valores nulos reales en campos textuales sin registro
- estabilizar la versión definitiva del dataset previo a su carga final en base de datos relacional

Con esta intervención se busca consolidar una tabla final limpia y consistente para las etapas de consulta SQL y modelado en Power BI.

```python
import pandas as pd
import re
import unicodedata

def normalizar_nombre_columna(col):
    col = str(col).strip().lower()
    col = unicodedata.normalize("NFKD", col).encode("ascii", "ignore").decode("utf-8")
    col = re.sub(r"\s+", "_", col)
    col = re.sub(r"[^a-z0-9_]", "_", col)
    col = re.sub(r"_+", "_", col)
    return col.strip("_")

# =============================
# 1. Eliminar columna duplicada primer_autor_inta(si,_no)
# =============================
cols_eliminar = []

for col in df_total.columns:
    col_norm = normalizar_nombre_columna(col)

    if (
        "primer_autor_inta" in col_norm
        and "si" in col_norm
        and "no" in col_norm
    ):
        cols_eliminar.append(col)

df_total = df_total.drop(columns=cols_eliminar, errors="ignore")


# =============================
# 2. Limpiar 0 en academico_inta_corresponding
# =============================
for col in df_total.columns:
    col_norm = normalizar_nombre_columna(col)

    if col_norm in [
        "academico_inta_corresponding",
        "academicos_inta_corresponding",
        "academico_inta_correspondiente",
        "academicos_inta_correspondientes",
        "academicos_inta_correspondiente"
    ]:
        df_total[col] = df_total[col].replace({
            0: pd.NA,
            0.0: pd.NA,
            "0": pd.NA,
            "0.0": pd.NA,
            "": pd.NA,
            "nan": pd.NA,
            "none": pd.NA,
            "None": pd.NA
        })

# =============================
# 3. Auditoría rápida
# =============================
print("Columnas eliminadas:")
print(cols_eliminar)

print("\nColumnas relacionadas con primer_autor:")
print([col for col in df_total.columns if "primer_autor" in normalizar_nombre_columna(col)])

print("\nColumnas relacionadas con corresponding:")
print([col for col in df_total.columns if "correspond" in normalizar_nombre_columna(col)])

print("\nShape final:")
print(df_total.shape)
```

```python
##FUNCION PARA NORMALIZAR HOJAS
def estandarizar(df):

    # normalizar columnas
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(' ', '_')
        .str.replace('á','a')
        .str.replace('é','e')
        .str.replace('í','i')
        .str.replace('ó','o')
        .str.replace('ú','u')
    )

    # unificar títulos
    if 'titulo_documento' in df.columns:
        df = df.rename(columns={'titulo_documento': 'titulo'})

    if 'titulo_de_documento' in df.columns:
        df = df.rename(columns={'titulo_de_documento': 'titulo'})

    # unificar if_5y
    if 'if5_y' in df.columns:
        df = df.rename(columns={'if5_y': 'if_5y'})

    if '5yif' in df.columns:
        df = df.rename(columns={'5yif': 'if_5y'})

    # unificar autor bool
    if 'primer_autor_inta_(si/no)' in df.columns:
        df = df.rename(columns={'primer_autor_inta_(si/no)': 'primer_autor_inta'})

    if 'primer_autor_inta_(si,_no)' in df.columns:
        df = df.rename(columns={'primer_autor_inta_(si,_no)': 'primer_autor_inta'})

    return df
```

```python
##Limpiamos de nuevo, ya que habian hojas conflictivas
hojas = [
    '2015','2016','2017','2018','2019',
    '2020','2021','2022','2023','2024','2025'
]

dfs = []

for h in hojas:
    df_temp = pd.read_excel(ruta, sheet_name=h)

    # Normalizar nombres de columnas
    df_temp.columns = (
        df_temp.columns
        .str.strip()
        .str.lower()
        .str.replace(' ', '_')
    )

    # Eliminar columnas basura tipo unnamed
    df_temp = df_temp.loc[:, ~df_temp.columns.str.contains('unnamed', case=False)]

    # Unificar nombres de columnas principales
    df_temp = df_temp.rename(columns={
        'revista': 'titulo_revista',
        'título_documento': 'titulo',
        'titulo_documento': 'titulo',
        'título_de_documento': 'titulo',
        'título': 'titulo'
    })

    # Caso especial hoja 2020
    if h == '2020':
        df_temp = df_temp.iloc[:146].copy()
        df_temp = df_temp[df_temp['titulo'].notna()].copy()

    # Año correcto
    df_temp['anio'] = int(h)

    dfs.append(df_temp)
```

```python
##Validamos el dataframe
for i, df in enumerate(dfs):
    print(f"{hojas[i]} -> {df.shape}")
```

```python
##Ahora unimos
df_total = pd.concat(dfs, ignore_index=True)

print(df_total.shape)
print(df_total.columns)
```

```python
##Tenemos filas duplicadas o con nombres no normalizados
# Unificar nombres de columnas después del concat
df_total = df_total.rename(columns={
    'título_revista': 'titulo_revista',
    '5yif': 'if_5y',
    'if5_y': 'if_5y',
    'if2022': 'if',
    'if2023': 'if',
    'if_2023': 'if',
    'primer_autor_inta_(si/no)': 'primer_autor_inta',
    'primer_autor_inta_(sí,_no)': 'primer_autor_inta',
    'prmer_autor_inta_(sí,_no)': 'primer_autor_inta',
    'acadèmico_inta_corresponding': 'academico_inta_corresponding'
})
```

```python
# Consolidar columnas duplicadas con el mismo nombre.
# Se transpone el DataFrame para agrupar por nombres de columnas,
# se toma el primer valor no nulo y luego se vuelve a transponer.
df_total = df_total.T.groupby(level=0).first().T
```

```python
##Validamos
# Verificar si aún quedan columnas duplicadas
print(df_total.columns[df_total.columns.duplicated()])

# Ver tamaño del dataset
print(df_total.shape)

# Ver columnas finales
print(df_total.columns)
```

```python
# Eliminar columna 'año' original del Excel.
# Usaremos 'anio', que viene desde el nombre de cada hoja y es más confiable.
df_total = df_total.drop(columns=['año'], errors='ignore')

# Validar cantidad de registros por año
print(df_total['anio'].value_counts().sort_index())
```

```python
# Detectar todas las columnas relacionadas con disciplina
cols_disc = [col for col in df_total.columns if 'disciplina' in col]

print(cols_disc)
```

```python
# Convertir columnas de disciplina a formato largo
df_total = df_total.melt(
    id_vars=[col for col in df_total.columns if col not in cols_disc],
    value_vars=cols_disc,
    var_name='disciplina_origen',
    value_name='disciplina_valor'
)

print(df_total.shape)
print(df_total[['titulo', 'disciplina_origen', 'disciplina_valor']].head(10))
```

```python
# Eliminar filas sin disciplina
df_total = df_total[df_total['disciplina_valor'].notna()].copy()

# Limpiar texto de disciplina
df_total['disciplina_valor'] = (
    df_total['disciplina_valor']
    .astype(str)
    .str.strip()
    .str.lower()
)

# Eliminar valores vacíos si quedaron
df_total = df_total[df_total['disciplina_valor'] != '']

# Renombrar columna final
df_total = df_total.rename(columns={'disciplina_valor': 'disciplina'})

print(df_total.shape)
print(df_total[['titulo', 'disciplina']].head(15))
```

```python
##Antes de exportar, analizaremos algunos datos
#Analizamos nulos por columnas
print(df_total.isnull().sum())

#todos los años
print(df_total['anio'].value_counts().sort_index())

#disciplina top
print(df_total['disciplina'].value_counts().head(15))
```

```python
##Limpieza final
# Limpiar texto general en columnas clave
cols_texto = [
    'primer_autor_inta',
    'academico_inta_corresponding',
    'coautor_inta',
    'q',
    'titulo_revista',
    'tipo_documento'
]

for col in cols_texto:
    if col in df_total.columns:
        df_total[col] = (
            df_total[col]
            .astype(str)
            .str.strip()
        )

# Normalizar primer_autor_inta
df_total['primer_autor_inta'] = (
    df_total['primer_autor_inta']
    .replace({
        'sí': 'Si',
        'si': 'Si',
        'Sí': 'Si',
        'no': 'No',
        '0': 'No',
        'nan': None
    })
)

# Limpiar Q
df_total['q'] = df_total['q'].astype(str).str.replace('\t', '').str.strip()
df_total['q'] = df_total['q'].replace('nan', None)

print(df_total[['primer_autor_inta','q']].head(20))
```

```python
##Estan todos los años
import pandas as pd

ruta_limpio = r'C:\Users\Nicolas Pozo\analisis-produccion-cientifica-inta\data\clean\publicaciones_inta_limpio.xlsx'
df = pd.read_excel(ruta_limpio)

print(df['anio'].value_counts().sort_index())
```

```python
##Titulos unicos
print("Filas totales:", len(df))
print("Titulos únicos:", df['titulo'].nunique())
```

```python
##Hay titulos vacios
print(df['titulo'].isnull().sum())
print((df['titulo'].astype(str).str.strip() == '').sum())
```

```python
##Las disciplinas se expandieron bien?
ejemplo = df[df['titulo'].str.contains('A simple molecular method', case=False, na=False)]
print(ejemplo[['titulo','disciplina']])
```
